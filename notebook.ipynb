```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- Config ---
CSV_PATH = "samples.csv"  # change if needed
NSHOW = -1  # number of samples to plot

# --- Load RX only ---
df = pd.read_csv(CSV_PATH)
n = df["n"].values[:NSHOW] if "n" in df.columns else np.arange(min(NSHOW, len(df)))
rx = df["rx_i"].values[:NSHOW].astype(float) + 1j * df["rx_q"].values[:NSHOW].astype(
    float
)

# --- Magnitude & Phase ---
mag = np.abs(rx)
ph_deg = np.degrees(np.unwrap(np.angle(rx)))  # unwrapped phase in degrees

# --- Plots ---
plt.figure()
plt.scatter(rx.real, rx.imag)
plt.title("Constellation")
plt.xlabel("I")
plt.ylabel("Q")
plt.show()

plt.figure()
plt.plot(n, mag)
plt.title("RX Magnitude")
plt.xlabel("Sample")
plt.ylabel("|RX|")
plt.show()

plt.figure()
plt.plot(n, ph_deg)
plt.title("RX Phase (unwrapped, deg)")
plt.xlabel("Sample")
plt.ylabel("Phase (deg)")
plt.show()
```
